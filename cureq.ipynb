# cureq

### Load in cure code and sample data

```q
\l cure.q
```

```q
dsc:{[n;r;s](r+n?s-r)*/:(cos;sin)@\:4*n?acos 0}
data :flip(-5 1)+(1 1.5)*dsc[1000; 0;1.8],'dsc[2000;3.1;4.2],'dsc[4000;5.2;6.5]
data,:flip(4 -1)+(1 8)*dsc[4000; 0;1.]
data@:neg[1000&count data]?count data
```

```q
plt:.p.import[`matplotlib]`:pyplot;
plt[`:scatter][;]. flip data;
plt[`:show][];
```

### Run the cure clustering algorithm


cureClust is in the form[data;numRep;comp;numClust] where:


data: The dataset to be clustered
numRep: Number of desired representative points
comp:compression applied to rep points
numClust:Number of desired clusters

```q
r:cureClust[data;20;0;4]
```

### Plot the clustered data

```q
plt:.p.import[`matplotlib]`:pyplot;
{plt[`:scatter][;]. flip x`pts}each r;
plt[`:show][];
```
